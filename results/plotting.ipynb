```python
df_list = []
depth = [2,4,6,8,10,12,14]
for i in depth:
    df = pd.read_pickle(f"hea_ry_100_q_{i}_layers_rand.pkl")
    df['depth'] = i
    df_list.append(df)
df_100_8 = pd.concat(df_list, ignore_index=False)
df_100_8.to_pickle('hea_ry_100q_8gpus')
ax = df_100_8.plot(x='depth', y='total_time', kind='scatter', logy=True, title='100Q Time vs. Depth 8GPU')
ax.set_xlabel("Layers")
ax.set_ylabel("Time (s)")
ax.plot()
```
